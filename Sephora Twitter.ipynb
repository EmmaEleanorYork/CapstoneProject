```python
name = "Sephora"
# Number of tweets to pull
tweetCount = 200

# Calling the user_timeline function with our parameters
results = api.user_timeline(id=name, count=tweetCount, include_rts = False)
```

```python
import pandas as pd
sephora_data = pd.DataFrame(data=[tweet.text for tweet in results],
                           columns = ['Tweets'])
```

```python
sephora_data.head()
```

```python
print(results[0].id)
print(results[0].created_at)
print(results[0].source)
print(results[0].favorite_count)
print(results[0].retweet_count)
print(results[0].entities)
```

```python
import numpy as np

sephora_data['ID'] = np.array([tweet.id for tweet in results])
sephora_data['Date'] = np.array([tweet.created_at for tweet in results])
sephora_data['Source'] = np.array([tweet.source for tweet in results])
sephora_data['Likes'] = np.array([tweet.favorite_count for tweet in results])
sephora_data['RTs'] = np.array([tweet.retweet_count for tweet in results])
sephora_data['Length'] = np.array([len(tweet.text) for tweet in results])
```
